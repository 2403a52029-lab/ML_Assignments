```python
import numpy as np

# Select score columns and convert to NumPy array
scores_np = df[['math_score', 'science_score', 'english_score', 'overall_score']].to_numpy()

print(scores_np)
```

```python
mean_score = np.mean(scores_np)
median_score = np.median(scores_np)
std_score = np.std(scores_np)

print("Mean:", mean_score)
print("Median:", median_score)
print("Standard Deviation:", std_score)
```

```python
mean_subject = np.mean(scores_np, axis=0)
median_subject = np.median(scores_np, axis=0)
std_subject = np.std(scores_np, axis=0)

print("Mean (Math, Science, English, Overall):", mean_subject)
print("Median:", median_subject)
print("Standard Deviation:", std_subject)
```

```python
min_score = np.min(scores_np)
max_score = np.max(scores_np)

print("Minimum score:", min_score)
print("Maximum score:", max_score)
```

```python
min_subject = np.min(scores_np, axis=0)
max_subject = np.max(scores_np, axis=0)

print("Minimum per subject:", min_subject)
print("Maximum per subject:", max_subject)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.hist(df['math_score'], bins=10)
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Math Scores")
plt.show()
```

```python
subjects = ['Math', 'Science', 'English']
average_scores = [
    df['math_score'].mean(),
    df['science_score'].mean(),
    df['english_score'].mean()
]

plt.figure()
plt.bar(subjects, average_scores)
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.title("Average Scores by Subject")
plt.show()
```

```python
plt.figure()
plt.scatter(df['math_score'], df['english_score'])
plt.xlabel("Math Score")
plt.ylabel("English Score")
plt.title("Math vs English Scores")
plt.show()
```

```python
gender_group = df.groupby('gender')[['math_score', 'science_score', 'english_score']].mean()
print(gender_group)
```

```python
parent_edu_group = df.groupby('parent_education')[['math_score', 'science_score', 'english_score']].mean()
print(parent_edu_group)
```

```python
gender_group.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Average Scores by Gender")
plt.show()
```

```python
parent_edu_group.plot(kind='bar')
plt.xlabel("Parental Education")
plt.ylabel("Average Score")
plt.title("Average Scores by Parental Education")
plt.show()
```
