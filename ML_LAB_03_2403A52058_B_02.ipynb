```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Salary_dataset.csv")

```

```python
print("First 5 rows:")
df.head()
```

```python
print("Last 5 rows:")
df.tail()
```

```python
print("Columns:", df.columns)
```

```python
X = df['YearsExperience']
y = df['Salary']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

print(X.shape, y.shape)
```

```python
m = 0.0  # slope
b = 0.0  # intercept
```

```python
def predict(X, m, b):
    return m * X + b
```

```python
def cost_function(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
y_pred = predict(X, m, b)
cost = cost_function(y, y_pred)
cost
```

```python
learning_rate = 0.01
epochs = 1000
n = len(X)

for i in range(epochs):
    y_pred = predict(X, m, b)

    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db

print("Trained slope (m):", m)
print("Trained intercept (b):", b)


```

```python
final_predictions = predict(X, m, b)
mse = cost_function(y, final_predictions)
mse
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, final_predictions, color='red', label='Regression Line')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
print(f"For every 1 year increase in experience, salary increases by approx {m:.2f}")
print(f"Base salary when experience is 0 years \u2248 {b:.2f}")
print(f"Mean Squared Error of the model: {mse:.2f}")
```
