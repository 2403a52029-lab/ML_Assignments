Objective:

To study relationships between variables using basic exploration.

Tasks:
1. Compute correlation matrix.
2. Identify highly correlated variables.
3. Plot scatter plot between age and maximum heart rate.
4. Interpret relationships.

Conclusion:
Relationships between medical parameters were explored.


```python
# 1. Compute correlation matrix
# Encode 'Heart Disease' column to numerical values
df_encoded = df.copy()
df_encoded['Heart Disease'] = df_encoded['Heart Disease'].map({'Absence': 0, 'Presence': 1})

correlation_matrix = df_encoded.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# 2. Identify highly correlated variables (absolute correlation > 0.7)
high_corr = correlation_matrix[abs(correlation_matrix) > 0.7]
print("\nHighly Correlated Variables:")
print(high_corr)
```

```python
# 3. Scatter plot between age and maximum heart rate (thalach)
plt.figure()
plt.scatter(df['Age'], df['Max HR'])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.title("Age vs Maximum Heart Rate")
plt.show()
```
